### measuring error in a neural network prediction

```python
knob_weight = 0.5 
input = 0.5 
goal_pred = 0.8 # real world observation

pred = input * knob_weight
#errror 
error = (pred - goal_pred) **2 # squaring forces output to be +ve

print (error)
```

### Neural learning(somple hot and cold method)

```python
#empty network 
weight = 0.1
lr = 0.01

def neural_network(input, weight):
    prediction = input * weight
    return prediction
 
#input data
number_of_toes = [8.5]
win_or_loss_binary = [1] #won

input = number_of_toes[0]
true = win_or_loss_binary[0] 
#make prediction 
pred = neural_network(input, weight)

#eval error using mean squared method
error =(pred - true) ** 2

print(error)
```

```python
lr = 0.1

#making prediction with an increased weight and eval error  
prediction_up = neural_network(input, weight+lr)
print("prediction with increased weight:",prediction_up)

error_up = (prediction_up - true) ** 2
print("error with increased weight", error_up)

print("------------")
#making prediction with a reduced weight and error eval 
prediction_down = neural_network(input , weight-lr)
print("prediction with reduced weight:",prediction_down)

error_down = (prediction_down - true)** 2
print("error with reduced weight", error_down)
```

```python
#comparing errors and adjusting weight
if (error > error_down or error > error_up):
    if(error_down < error_up):
        weight -= lr
        
    if(error_up > error_down):
        weight += lr
```

## Hot and Cold Learning

```python
weight = 0.5
input = 0.5
goal_prediction = 0.8
step_amount = 0.001 #amount to adjust weight by at each iteration

for iteration in range(1102): #range = to times to to repeat learning - repeating more leads to a smaller error
    prediction = input * weight
    
    error = (prediction - goal_prediction)** 2
    print("Error: " + str(error) + " prediction:" + str(prediction))
    
    #prediction with inreased weight
    pred_weight_up = input * (weight + step_amount)
    w_up_error = (goal_prediction - pred_weight_up)**2
    
    #prediction with inreased weight
    pred_weight_down = input * (weight - step_amount)
    w_down_error = (goal_prediction - pred_weight_down)**2
    
    #compare errors and adjusting weight
    
    if(w_down_error < w_up_error):
        weight = weight - step_amount
    
    if(w_up_error < w_down_error):
        weight = weight + step_amount
    
```
